```python
#Crea una lista con 5 nomi di amici e stampa:il primo nome    l’ultimo nome    la lunghezza della lista
nomi=['edo','giuseppe','antonio','franco','pio']
print(nomi[0])
print(nomi[-1])
print(len(nomi))
#Aggiungi un nuovo nome alla lista con .append().
nomi.append('franco')
print(nomi)
#Rimuovi un nome dalla lista con .remove().
nomi.remove('antonio')
print(nomi)
#Ordina la lista in ordine alfabetico con .sort() e stampala.
nomi.sort()
print(nomi)
```
