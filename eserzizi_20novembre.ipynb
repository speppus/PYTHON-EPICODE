```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(override=True, dotenv_path="file_connessione_db_AdvanterWork.env")

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")


connection_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
print(username, password, host, dbname)
db_engine = sqlalchemy.create_engine(connection_string)

query = "SELECT * FROM dimproduct"
pd.read_sql(query, db_engine)

#valutiamo quanto segue: • Quanti dati ci sono in totale? • Quali sono i metadati? • Stampiamo il primo elemento • Stampiamo l'ultimo elemento • Riusciamo a stampare cinque elementi a caso? • Quali sono i colori disponibili?
pd=pd.read_sql(query, db_engine)
print(pd.shape)  #Quanti dati ci sono in totale?
print(pd.columns)#Quali sono i metadati?
print(pd.head(1))#Stampiamo il primo elemento
print(pd.tail(1))#Stampiamo l'ultimo elemento
print(pd.sample(5))#Riusciamo a stampare cinque elementi a caso?
print(pd['Color'].value_counts())#Quali sono i colori disponibili? e quanti?
```

```python
#In media quanto pesano i prodotti (per questo esercizio, ignoriamo l'unità di misura, ma usiamo solo i valori della colonna relativa)? 
# • Quanto pesa il più leggero? • Quanto pesa il più pesante? • Quanti prodotti pesano più di 100? • Quanto costano in media i prodotti (colonna DealerPrice)? • 
# Quali sono i quartili dei prezzi?
print(pd['Weight'].min())#Quanto pesa il più leggero?
print(pd['Weight'].mean())#Quanto pesa il più pesante?
filtro=pd['Weight']>100#Quanti prodotti pesano più di 100? 
print(pd.loc[filtro])#Quanti prodotti pesano più di 100? 
print(pd['DealerPrice'].mean())#Quanto costano in media i prodotti (colonna DealerPrice)? • 
print(pd['DealerPrice'].quantile(0.25))# Quali sono i quartili dei prezzi? 1° quantile
print(pd['DealerPrice'].quantile(0.50))# Quali sono i quartili dei prezzi?  2° quantile
print(pd['DealerPrice'].quantile(0.75))# Quali sono i quartili dei prezzi?   3° quantile
```

```python
#• Qual è il prezzo medio per i prodotti di colore blu? • Qual è il prezzo medio per i prodotti di colore rosso o nero? •
#  Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? • 
# Qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento? •
#  Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: quali sono le differenze? C'è un pattern? C'è qualche elemento che non lo segue?

filtrocolore=pd.loc[pd['Color']=='Blue','DealerPrice'].mean()
print(filtrocolore) #• Qual è il prezzo medio per i prodotti di colore blu?
filtrocolori=pd.loc[(pd['Color']=='Black') | (['Color']== 'Red'),'DealerPrice'].mean()
print(filtrocolori)

filtrotaglia=pd['Size']=='42'#Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg? • 
filtropeso=pd['Weight']>10 
filtro=pd.loc[filtrotaglia & filtropeso]
prezzomassimo=filtro['DealerPrice'].max()
print(prezzomassimo)

filtrodef=(pd['Size']=='42') & (pd['Weight']>10) & (pd['Color']=='Silver')
risultato=pd.loc[filtrodef,['StandardCost', 'EnglishProductName']]
print(risultato)

ultimi_20 = pd.tail(20)
print(ultimi_20[["StandardCost", "DealerPrice"]])#  Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset



```
