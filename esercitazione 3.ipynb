```python
#L'azienda Object SpA ha creato una lista di quanti oggetti ha venduto ogni mese nell'ultimo anno:
#  lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350] trasformiamola in un array NumPy (casting): 
# lst = np.array(lst) e rispondiamo alle domande del CEO della Object SpA: •
#  qual è stata la vendita massima mensile? E quella minima? • 
# quali sono le vendite mensili maggiori di 5000 oggetti? E quante ne sono? •
#  quali sono le vendite minori di 3000 oggetti? 5 • in media quanti oggetti sono stati venduti al mese?
import numpy as np
lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350]
lst=np.array(lst)
print(np.max(lst))
print(np.min(lst))
print(lst>5000)
filtro=lst>5000
print(len(lst[filtro]))
print(lst<3000)
vendite_minori=lst<3000
print(len(lst[vendite_minori]))
print(np.mean(lst))

```

```python
#Consideriamo il seguente dizionario: 
# fatturati_dict = {1997: 12_000, 1998: 15_000, 1999: 20_000, 2000: 23_000, 2001: 25_000, 2002: 17_000, 2003: 14_000, 2004: 21_000} 
# Consideriamo ora la seguente Series: fatturati_series = pd.Series([12_000, 15_000, 20_000, 23_000, 25_000, 17_000, 14_000, 21_000], index=range(1997, 2005)) 
# Possiamo accedere alle stesse informazioni nello stesso modo: fatturati_dict[1997] fatturati_series[1997] Dunque qual è la differenza tra i due tipi di dato?
#  6 Cosa potremmo fare con la Series che non possiamo fare con il dizionario?
fatturati_dict = {1997: 12_000, 1998: 15_000, 1999: 20_000, 2000: 23_000, 2001: 25_000, 2002: 17_000, 2003: 14_000, 2004: 21_000}
fatturati_series = pd.Series([12_000, 15_000, 20_000, 23_000, 25_000, 17_000, 14_000, 21_000], index=range(1997, 2005))

print(fatturati_dict.keys())  #questa formula permette di vedere le chiavi del dizionario
print(fatturati_dict.values())  #questa formula permette di vedere i valori del dizionario
print(fatturati_dict.items()) #questa formula permette di vedere entrrambi chiavi e valori
print(fatturati_dict[1997])  #con questo filtro per posizione possiamo vedere solo il valore senza effettuare operazioni matematiche
print(fatturati_series[1997]) #invece con la serie e possibile oltre a visualizzzare il valore anche a lavorarci sopra 
print(fatturati_series[1997+1]) # ad esempio ho filtrato anno 1998 aggiungendo 1 anno
```

```python
#Abbiamo una lista di stringhe di prezzi in dollari, che erroneamente sono stati scritti con il simbolo dell'euro:
#  prezzi = ["100 €", "200 €", "500 €", "10 €", "50 €", "70 €"] 
# cambiare il simbolo dell'euro (€) in quello del dollaro ($) per ogni stringa nella lista; il risultato sarà memorizzato in un'altra lista.

prezzi = ["100 €", "200 €", "500 €", "10 €", "50 €", "70 €"] 
prezziDollaro=[]

for x in prezzi:
     nuoviprezzi=x.replace('€','$')
     prezziDollaro.append(nuoviprezzi)
        
print(prezziDollaro)

                       
```
